```python
df = pd.read_excel('train_sprocket.xlsx')
```

```python
df.head()
```

```python
Q1 = df['customer_age'].quantile(0.25)
Q3 = df['customer_age'].quantile(0.75)
IQR = Q3 - Q1
lowet_value = Q1 - 1.5*IQR
highest_value = Q3 + 1.5*IQR
```

```python
df = df[(df['customer_age'] > lowet_value) & (df['customer_age'] < highest_value)]
df.shape
```

```python
df.to_excel('train_sprocket_.xlsx', index=None)
```
